```python
df['price'] = df['price'].astype('int64')
```

```python
numFamilyCars = df[(df['num of doors'] == 'four') & (df['price'] < 20000) & (df['city mpg'] > 17)].count().iloc[0]
```

```python
totalCars = df.count().iloc[0]
percentage = numFamilyCars / totalCars
print('total cars',totalCars)
```

```python
countGas  = int(df[df['fuel-type'] == 'gas'].count().iloc[0])
countDisel = int(df[df['fuel-type'] == 'diesel'].count().iloc[0])
print("number of gas cars:",countGas)
print("number of disel cars:",countDisel)
totalCars = df.count().iloc[0]
ratio = countDisel/countGas

print("the ratio of disel w.r.t to gas:",str(ratio))
```
